```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# adjusted timeline for thesis completion
tasks = [
  ("Setup local testing environment", "2025-09-15", "2025-12-22"),
  ("Hollidays and birthday", "2025-12-22", "2025-12-31"),
  ("Model - Anomaly Detection", "2026-01-01", "2026-01-31"),
  ("Model - Service Graph", "2026-02-01", "2026-02-28"),
  ("OpenWebUI & LLM setup", "2026-03-01", "2026-03-15"),
  ("MCP servers setup", "2026-03-16", "2026-04-15"),
  ("Finish writing thesis", "2026-04-16", "2026-05-31"),
]

colors = ["#A9A9A9", "#87CEEB", "#7CFC00", "#7CFC00", "#7CFC00", "#7CFC00", "#7CFC00"]

labels = [t[0] for t in tasks]
start = [datetime.strptime(t[1], "%Y-%m-%d") for t in tasks]
end = [datetime.strptime(t[2], "%Y-%m-%d") for t in tasks]
durations = [(e - s).days for s, e in zip(start, end)]

fig, ax = plt.subplots(figsize=(12,6))

# reverse so first task is at top
labels_rev = labels[::-1]
start_rev = start[::-1]
dur_rev = durations[::-1]
colors_rev = colors[::-1]

ax.barh(labels_rev, dur_rev, left=start_rev, edgecolor="black", color=colors_rev)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax.grid(True, axis='x', linestyle='--', alpha=0.6)
ax.set_title("Master's Thesis adjusted Timeline")

plt.tight_layout()
```
